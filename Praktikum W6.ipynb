```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
from sklearn import tree
import pandas as pd
import numpy as np
```

```python
# Assignment 1
dataset_train = pd.read_csv('milk_training.csv')
dataset_train
```

```python
# Assignment 2
train_data = dataset_train.iloc[:, :-1]
train_data
```

```python
# Assignment 3
dataset_test = pd.read_csv('milk_testing.csv')
dataset_test
```

```python
# Assignment 4
test_data = dataset_test.iloc[:, :-1]
test_data
```

```python
# Assignment 5
train_label = dataset_train.iloc[:, -1]
train_label
```

```python
# Assignment 6
test_label = dataset_test.iloc[:, -1]
test_label
```

```python
# Assignment 7
dtc = DecisionTreeClassifier()
dtc.fit(train_data, train_label)

class_result = dtc.predict(test_data)
print('Class = ', class_result)

acc = dtc.score(test_data, test_label)
err = 1 - acc
print("\n\nAkurasi ratio Decision Tree: ", acc)
print("Error ratio Decision Tree: ", err)
```

```python
# Assignment 8
plt.figure(figsize=(15, 20))
plot_tree(dtc, feature_names=train_data.columns.values, 
          class_names=[str(c) for c in train_label.unique()], 
          filled=True)
plt.show()
```

```python
# Assignment 9
acc = dtc.score(test_data, test_label)
err = 1 - acc
print("Akurasi ratio Decision Tree: ", acc)
print("Error ratio Decision Tree: ", err)

kNN = KNeighborsClassifier(n_neighbors=3, weights='distance')
kNN.fit(train_data, train_label)  
y_pred = kNN.predict(test_data)
accuracy = accuracy_score(test_label, y_pred)
error_ratio_knn = 1 - accuracy

print("\nAkurasi ratio KNN : ", accuracy)
print("Error ratio KNN : ", error_ratio_knn)

classifier = GaussianNB()
classifier.fit(train_data, train_label)
ypred = classifier.predict(test_data)
accuracy = accuracy_score(test_label, ypred)
error_ratio_bayesian = 1 - accuracy

print("\nAkurasi ratio Bayesian : ", accuracy)
print("Error ratio Bayesian : ", error_ratio_bayesian)
```
